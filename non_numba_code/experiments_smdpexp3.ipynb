Plotting of SMDP-EXP3


```python
import numpy as np

import importlib


import matplotlib.pyplot as plt
import gc
import pandas as pd
```

Plots nS = 10 for different values of N

```python
# Loop plots:
# Lopp over N's
for n in [5,10,15,20]:
        # Read data
        smdpexp10_10 = pd.read_pickle("C:/Users/andre/Bachelor_code/experiment_results/smdpexp3_rs/smpdexp3_ns_10_Tmax_5_N"+str(n)+".pkl")
        smdpexp10_15 = pd.read_pickle("C:/Users/andre/Bachelor_code/experiment_results/smdpexp3_rs/smpdexp3_ns_10_Tmax_10_N"+str(n)+".pkl")
        smdpexp10_20 = pd.read_pickle("C:/Users/andre/Bachelor_code/experiment_results/smdpexp3_rs/smpdexp3_ns_10_Tmax_15_N"+str(n)+".pkl")

        bus10 = pd.read_pickle("C:/Users/andre/Bachelor_code/experiment_results/bus_opt_rs/bus_ns_10_opt_H_100")
        smdpucrl_l =  pd.read_pickle("C:/Users/andre/Bachelor_code/experiment_results/rs_ucrl_l/nS_10_correct_confidence.pkl")
        smdpucrl_l10 = smdpucrl_l['UCRL_SMDP-L, T_max = 10']
        
        # plots
        plt.plot(smdpexp10_10,label='SMDP-EXP3, T_max = 5,  N = '+str(n))
        plt.plot(smdpexp10_15,label='SMDP-EXP3, T_max = 10, N = '+str(n))
        plt.plot(smdpexp10_20,label='SMDP-EXP3, T_max = 15, N = '+str(n))

        plt.plot(bus10,label='BUS, H = 100')
        plt.plot(smdpucrl_l10,label='SMDP-UCRL, T_max = 10')
        plt.xlabel('T')
        plt.ylabel('Cumulative Regret')
        plt.title('nS = 10')

        plt.legend()
        plt.legend(loc='center', framealpha = 0.05, bbox_to_anchor = (0.5,-0.25), ncol = 3 ,prop={'size': 7.5})
        plt.tight_layout()
        plt.grid()
        plt.savefig("C:/Users/andre/Bachelor_code/smdpexp3_nS10_N"+str(n)+".png")
        plt.show()
```
