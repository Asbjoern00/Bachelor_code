```python
# Generate T_max data (i.e. create data with the similar ones in one)

for i in range(2,8):    
    b_data = binom[['UCRL_SMDP, T_max = '+str(i)]]
    b_data.rename({'UCRL_SMDP, T_max = '+str(i): 'UCRL_SMDP, binomial, T_max = '+str(i)}, axis=1, inplace=True)

    b_data= pd.concat([b_data, unif[['UCRL_SMDP, T_max = '+str(i)]]], axis=1)
    b_data.rename({'UCRL_SMDP, T_max = '+str(i): 'UCRL_SMDP, uniform, T_max = '+str(i)}, axis=1, inplace=True)

    bl_data = binom[['UCRL_SMDP-L, T_max = '+str(i)]]
    bl_data.rename({'UCRL_SMDP-L, T_max = '+str(i): 'UCRL_SMDP-L, binomial, T_max = '+str(i)}, axis=1, inplace=True)

    bl_data= pd.concat([bl_data, unif[['UCRL_SMDP-L, T_max = '+str(i)]]], axis=1)
    bl_data.rename({'UCRL_SMDP-L, T_max = '+str(i): 'UCRL_SMDP-L, uniform, T_max = '+str(i)}, axis=1, inplace=True)

    # plots
    utils.plot_mean_regret_from_pd(b_data,8)
    plt.legend(framealpha=0.05,loc='center', bbox_to_anchor=(0.5, -0.25), ncol=2,prop={'size': 8})
    plt.tight_layout()
    plt.savefig('C:/Users/andre/OneDrive/KU, MAT-OEK/Bachelor matematik-oekonomi/3. aar/Bachelorprojekt/Out/distr_Tmax'+str(i)+'_nS8.png')
    plt.show()
    utils.plot_mean_regret_from_pd(bl_data,8)
    plt.legend(framealpha=0.05,loc='center', bbox_to_anchor=(0.5, -0.25), ncol=2,prop={'size': 8})
    plt.tight_layout()
    plt.savefig('C:/Users/andre/OneDrive/KU, MAT-OEK/Bachelor matematik-oekonomi/3. aar/Bachelorprojekt/Out/distr_Tmax'+str(i)+'_nS8_L.png')
    plt.show()
```

```python
# Plot of expected holdingtimes going right. 
def ratio(T_max,p):
    return ((T_max+1)/2)/(p*T_max+(1-p)**T_max)

plt.plot(range(1,9),[ratio(t,0.9) for t in range(1,9)],label = 'Ratio')

plt.title(r'Ratio of expected holding time')
plt.ylabel(r'$\bar{\tau}_{unif}(s,R)/\bar{\tau}_{binom}(s,R)$')
plt.xlabel(r'$T_{max}$')
plt.legend(framealpha=0.05,loc='center', bbox_to_anchor=(0.5, -0.25), ncol=1,prop={'size': 8})
plt.tight_layout()
plt.grid()
plt.savefig('C:/Users/andre/OneDrive/KU, MAT-OEK/Bachelor matematik-oekonomi/3. aar/Bachelorprojekt/Out/ratio_unif_binom.png')
```
